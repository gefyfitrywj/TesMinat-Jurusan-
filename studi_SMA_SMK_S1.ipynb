```python
from sklearn.model_selection import train_test_split

# Pemisahan fitur (X) dan target (y)
X = data.iloc[:, :-2]  # Semua kolom kecuali dua target terakhir
y_sma = data["Jurusan SMA/SMK"]
y_s1 = data["Program Studi (Sarjana)"]
```

```python
# Membagi dataset menjadi training (80%) dan testing (20%)
X_train_sma, X_test_sma, y_train_sma, y_test_sma = train_test_split(X, y_sma, test_size=0.2, random_state=42)
X_train_s1, X_test_s1, y_train_s1, y_test_s1 = train_test_split(X, y_s1, test_size=0.2, random_state=42)

print("\nUkuran Data Train & Test:")
print(f"SMA/SMK - Train: {X_train_sma.shape}, Test: {X_test_sma.shape}")
print(f"S1 - Train: {X_train_s1.shape}, Test: {X_test_s1.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Melatih model Random Forest untuk SMA/SMK
rf_sma = RandomForestClassifier(n_estimators=100, random_state=42)
rf_sma.fit(X_train_sma, y_train_sma)
y_pred_sma = rf_sma.predict(X_test_sma)

# Evaluasi model SMA/SMK
print("\nEvaluasi Model SMA/SMK:")
print(f"Akurasi: {accuracy_score(y_test_sma, y_pred_sma) * 100:.2f}%")
print(classification_report(y_test_sma, y_pred_sma))
print("Confusion Matrix:")
print(confusion_matrix(y_test_sma, y_pred_sma))

# Melatih model Random Forest untuk S1
rf_s1 = RandomForestClassifier(n_estimators=100, random_state=42)
rf_s1.fit(X_train_s1, y_train_s1)
y_pred_s1 = rf_s1.predict(X_test_s1)

# Evaluasi model S1
print("\nEvaluasi Model S1:")
print(f"Akurasi: {accuracy_score(y_test_s1, y_pred_s1) * 100:.2f}%")
print(classification_report(y_test_s1, y_pred_s1))
print("Confusion Matrix:")
print(confusion_matrix(y_test_s1, y_pred_s1))
```

Hasil Evaluasi Model

* SMA/SMK: Akurasi 99.29%, dengan beberapa kesalahan kecil pada kategori SMA IPS, IPA
* S1: Akurasi 95.05%, dengan beberapa kesalahan pada S1 Sastra Hindi dan S1 Sastra Inggris



## Model Uji (Cross-Validation)

```python
from sklearn.model_selection import cross_val_score

# Assuming you want to use the Random Forest model for SMA/SMK (rf_sma)
rf_model = rf_sma  # Assign rf_sma to rf_model
y_encoded = y_sma  # Assign y_sma to y_encoded

# Cross-validation untuk Random Forest
cv_scores_rf = cross_val_score(rf_model, X, y_encoded, cv=5)
print("Akurasi Cross-Validation Random Forest:", cv_scores_rf)
print("Rata-rata Akurasi:", np.mean(cv_scores_rf))

# Cross-validation untuk Program Studi Sarjana (S1)
rf_model_s1 = rf_s1  # Model Random Forest untuk S1
y_encoded_s1 = y_s1  # Label untuk S1

cv_scores_rf_s1 = cross_val_score(rf_model_s1, X, y_encoded_s1, cv=5)
print("\nAkurasi Cross-Validation Random Forest untuk Program Studi Sarjana (S1):", cv_scores_rf_s1)
print("Rata-rata Akurasi S1:", np.mean(cv_scores_rf_s1))
```

Berdasarkan hasil cross-validation, model Random Forest untuk rekomendasi SMA/SMK memiliki rata-rata akurasi 99.43%, yang sangat tinggi dan stabil di setiap lipatan validasi.

Model SMA/SMK
* Akurasi pengujian mencapai 99.29%, menunjukkan bahwa model memiliki tingkat prediksi yang sangat akurat.
* Hasil cross-validation dengan Random Forest menunjukkan rata-rata akurasi 99.43%, yang mengindikasikan model memiliki performa yang stabil di berbagai lipatan data uji.

Model S1
* Akurasi pengujian sebesar 95.05%, yang tetap menunjukkan performa yang baik meskipun sedikit lebih rendah dibandingkan model SMA/SMK.
* Perbedaan ini mungkin disebabkan oleh kompleksitas data atau variasi lebih besar dalam preferensi jurusan di tingkat S1.
